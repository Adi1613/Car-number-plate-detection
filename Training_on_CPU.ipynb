## Import dependencies 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from darkflow.net.build import TFNet
import cv2
```

```python
options = {"model": "ckpt/cfg/tiny_yolo_custom.cfg", 
           "load": "bin/tiny-yolo-voc.weights",
           "batch": 8,
           "epoch": 100,
           "gpu": 0.0,
           "train": True,
           "annotation": "./adi/train/annotation/",
           "dataset": "./adi/train/images/"}
```

```python
tfnet = TFNet(options)
```

## Training the Model

```python
tfnet.train()
```

```python
# this line of code lets you save the built graph to a protobuf file (.pb)
# this step is unnecessary for this notebook
tfnet.savepb()
```

## Loading from the Latest Checkpoint

This cell shows how to load custom trained model from the checkpoint. It will easy your work.

1. model: the model configuration that you used (*.cfg)

2. load: set -1. In this way, the latest checkpoint will be loaded.

```python
options = {"model": "ckpt/cfg/tiny_yolo_custom.cfg",
           "load": -1,
           "gpu": 0.0}
```

```python
tfnet2 = TFNet(options)
```

### load_from_ckpt()
By calling load_from_ckpt() method in TFNet, it will load pre-trained parameters from the checkpoint that you just specified in options.

```python
tfnet2.load_from_ckpt()
```

## Predicting on a Image

```python
from datetime import datetime
start_time = datetime.now()

import pprint as pp

original_img = cv2.imread("./adi/test/car_10.jpg")
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
results = tfnet2.return_predict(original_img)
print(results)

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
def boxing(original_img , results):
    newImage = np.copy(original_img)

    for result in results:
        top_x = result['topleft']['x']
        top_y = result['topleft']['y']

        btm_x = result['bottomright']['x']
        btm_y = result['bottomright']['y']

        confidence = result['confidence']
        label = result['label'] + " " + str(round(confidence, 3))
        
        if confidence > 0.3:
            newImage = cv2.rectangle(newImage, (top_x, top_y), (btm_x, btm_y), (255,0,0), 3)
            newImage = cv2.putText(newImage, label, (top_x, top_y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL , 0.8, (0, 230, 0), 1, cv2.LINE_AA)
        
    return newImage
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.imshow(boxing(original_img, results))
```

### Predicting on few test images with the time of execution

```python
start_time = datetime.now()
from math import ceil

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))

for i in range(5):
    original_img = cv2.imread("./adi/test/car_" + str(i+1) + ".jpg")
    original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
    results = tfnet2.return_predict(original_img)
    
    ax[ceil(i/3)-1, i%3].imshow(boxing(original_img, results))
    end_time = datetime.now()
    print('Duration: {}'.format(end_time - start_time))
```
